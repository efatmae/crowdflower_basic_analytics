```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df_CF_judgments = pd.read_csv("Turkish_Gezi_protest/CF_aggregate_best_answer_with_test_questions_22_05_2018.csv")
```

```python
len(df_CF_judgments)
```

```python
df_CF_judgments_test_questions = df_CF_judgments[df_CF_judgments['_golden']==True]
```

```python
df_CF_judgments_test_questions.columns
```

```python
df_CF_judgments_test_questions.rename(columns={'does_this_tweet_reportdiscuss_violent_incidents_happened_during_the_protest':'violence',
                         'is_this_tweet_related_to_turkish_protests_2013' : 'protest',
                         'does_this_tweet_reportdiscuss_violent_incidents_happened_during_the_protest:confidence':'violence_confidence',
                          'is_this_tweet_related_to_turkish_protests_2013:confidence':'protest_confidence'
                         }, inplace = True)
```

```python
df_CF_judgments_test_questions
```

```python
print('avg user confidence level for protest test question', np.mean(
   df_CF_judgments_test_questions['protest_confidence'])) 
```

```python
plt.hist(df_CF_judgments_test_questions['protest_confidence'],50, density=1)
plt.title('test questions protest confidence levels')
plt.show()
```

```python
print('avg user confidence level for violence test question', np.mean(
    df_CF_judgments_test_questions['violence_confidence'])) 
```

```python
x = df_CF_judgments_test_questions['violence_confidence'].dropna()
plt.hist(x,50, density=1)
plt.title('test questions violence confidence levels')
plt.show()
```
